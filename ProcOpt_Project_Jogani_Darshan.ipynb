### Gradient methods
#### a)   Which condition does the descent direction $v^{(k)}$ have to fulfill? Formulate it mathematically. When is this condition met for the gradient descent method? Show mathematically. **(4P)**

<span style="color:green"><font size="5"><b>Answer:</b></font></span> <br>

The descent direction $v^{(k)}$ must fulfill the condition that it produces a decrease in the function value at the current point $x^{(k)}$. Mathematically, this is expressed as:

\begin{equation*}
\Big(F_x^{(k)}\Big)^T v^{(k)} < 0
\end{equation*}

where;
- $\Big(F_x^{(k)}\Big)$ is the gradient of the function $F$ at the current point $x^{(k)}$, and
- $v^{(k)}$ is the descent direction.

This condition is met for the gradient descent method because the descent direction is chosen as the negative of the gradient.

Mathematically, this is expressed as:

\begin{equation*}
v^{(k)} = -\Big(F_x^{(k)}\Big)
\end{equation*}

Substituting this into the condition gives:

\begin{equation*}
\Big(F_x^{(k)}\Big)^T \Bigg(-\Big(F_x^{(k)}\Big)\Bigg) < 0
\end{equation*}

which simplifies to:

\begin{equation*}
-\left|\left|F_x^{(k)}\right|\right| ^2 < 0
\end{equation*}

Since the square of the norm of the gradient is always non-negative, this condition is always met for the gradient descent method (except at a stationary point where the gradient is zero).

#### b)   When using line search, formulate the subproblem which has to be solved in every iteration. Name the variables. Name two methods which can be used to solve it. Can the Nelder-Mead Simplex methods be used for this? Give the reason. **(6P)** 

<span style="color:green"><font size="5"><b>Answer:</b></font></span> <br>

In line search, the subproblem to be solved in every iteration is to find the step size that minimizes the objective function along the search direction.

Mathematically, this can be formulated as:

Given a descent direction $v^{(k)}$ at the current point $x^{(k)}$, find $\alpha^{(k)} > 0$ that minimizes $F(x^{(k)} + \alpha^{(k)} v^{(k)})$.

where:
- $F$ is the objective function,
- $x^{(k)}$ is the current point,
- $\alpha ^{(k)}$ is the step size (learning rate), and
- $v^{(k)}$ is the descent direction.

Two methods (Wolfe Conditions) that can be used to solve this subproblem are:
1. "Armijo" rule (sufficient descent): For a valid step, the function value needs to decrease sufficiently.
2. Curvature condition (sufficiently flat): For a valid step, the gradient of the function along the descent direction needs to decrease sufficiently.

No, the Nelder-Mead Simplex method cannot be used for line search. This is because the Nelder-Mead method is a direct search method that does not make use of gradient information, while line search methods typically require the use of gradients to determine the search direction and step size. Furthermore, the Nelder-Mead method operates in a simplex (a polytope of n+1 vertices in n dimensions), not along a line.

### Newton methods
#### c)   Formulate mathematically the iteration update to solve an optimization problem using the Newton-Raphson method. Which steps are combined in Newton-Raphson? For which type of problem does the method converge in one step? **(4P)**

<span style="color:green"><font size="5"><b>Answer:</b></font></span> <br>

The iteration update in the Newton-Raphson method for solving an optimization problem is formulated as follows:

\begin{equation*}
x^{(k+1)} = x^{(k)} + \Delta x^{(k)}
\end{equation*}

where;

\begin{equation*}
\Delta x^{(k)} = - [F_{xx}(x^{(k)})]^{-1} F_x(x^{(k)})
\end{equation*}

where:
- $x^{(k)}$ is the current point,
- $x^{(k+1)}$ is the next point,
- $F_{xx}(x^{(k)})$ is the Hessian matrix (matrix of second derivatives) of the function $F$ at the current point,
- $[F_{xx}(x^{(k)})]^{-1}$ is the inverse of the Hessian matrix, and
- $F_x(x^{(k)})$ is the gradient (vector of first derivatives) of the function $F$ at the current point.

The Newton-Raphson method combines two steps: 
1. It computes the direction of the step by using the inverse of the Hessian matrix to scale the gradient.
2. It then updates the current point by taking a step in the computed direction.

The Newton-Raphson method converges in one step for quadratic problems. This is because the second-order Taylor series approximation used by the method is exact for quadratic functions.

#### d)   Name three major problems which arise when using the Newton-Raphson method and explain how Quasi-Newton methods improve every problem. **(6P)**

<span style="color:green"><font size="5"><b>Answer:</b></font></span> <br>

1. **Computational Cost**: The Newton-Raphson method requires the computation and inversion of the Hessian matrix at each iteration, which can be computationally expensive.

   - While, Quasi-Newton methods, such as the Broyden-Fletcher-Goldfarb-Shanno (BFGS) method, improve this by approximating the Hessian matrix at each step instead of computing it directly. This significantly reduces the computational cost.

2. **Non-Positive Definite Hessian**: The Newton-Raphson method requires the Hessian to be positive definite for the method to converge to a minimum. If the Hessian is not positive definite, the method may not converge.

   - While, Quasi-Newton methods address this by ensuring that the approximate Hessian is always positive definite. For example, the BFGS method updates the Hessian approximation in a way that maintains positive definiteness.

3. **Inversion of the ill-conditioned Hessian**: The Newton-Raphson method can have issues with ill-conditioned matrices. An ill-conditioned Hessian matrix can lead to numerical instability and inaccurate results because small changes in the input can lead to large changes in the output. This is particularly problematic when inverting the Hessian.

   - While, Quasi-Newton methods typically work with an approximation to the inverse Hessian, rather than the Hessian itself. This means they do not need to invert a potentially ill-conditioned matrix, which can lead to numerical instability.

To summerize, Quasi-Newton methods offer a good balance between the fast convergence of the Newton-Raphson method and the robustness and lower computational cost of gradient descent methods.

## 3. Optimal operation of a methanation reactor (56P)

A biogas plant typically produces a mixture of CO<sub>2</sub> and CH<sub>4</sub>. The product gas is to be further processed in a continuously operating
ideal plug flow tube reactor in order to increase the methane content and feed it
into the natural gas network. The methanation reaction taking place in the
reactor

\begin{equation*}
    \mathrm{CO_2} + 4 \, \mathrm{H_2} \leftrightarrow \mathrm{CH_4} + 2 \, \mathrm{H_2O} 
\end{equation*}

converts excess carbon dioxide to methane and water. The reaction requires hydrogen, which can be supplied by water electrolysis. The reaction rate $r$ of this reaction can be calculated using the following equation:

\begin{equation}
    r = k \, p_{\mathrm{CO_2}}^{0.5} \, p_{\mathrm{H_2}}^{0.5} \, \left( 1 - \frac{p_{\mathrm{CH4}} \, p_{\mathrm{H_2O}}^2}{p_{\mathrm{CO_2}} \, p_{\mathrm{H_2}}^4 \, K_{\mathrm{eq}}}  \right) \, \frac{1}{Q^2}
\end{equation}

Where $p_\mathrm{i}$ are the partial pressures of the component $\mathrm{i}$ $\left(\mathrm{i} = \mathrm{CH_4}, \mathrm{H_2O}, \mathrm{H_2},\mathrm{CO_2} \right)$, $k$ denotes the (temperature dependent) rate constant:

\begin{equation*}
    k = k_0 \exp\left( \frac{E_\mathrm{A}}{R} \left( \frac{1}{T_{\mathrm{ref}}} - \frac{1}{T} \right) \right)
\end{equation*}

and $Q$ denotes the dependence of the reaction rate $r$ on the adsorption of the components on the catalyst:

\begin{equation*}
    Q = 1 + K_{\mathrm{mix}} \, p_{\mathrm{CO_2}}^{0.5} + K_\mathrm{H_2} \, p_{H_2}^{0.5} + \frac{K_\mathrm{OH} \, p_{\mathrm{H_2O}}}{p_{\mathrm{H_2}}^{0.5}}
\end{equation*}

The (temperature dependent) adsorption constants $K_{\mathrm{mix}}, K_{\mathrm{H_2}}, K_{\mathrm{OH}} $ can be calculated as follows:

\begin{align*}
K_{\mathrm{mix}} &= k_{0,\mathrm{mix}} \, \exp\left( \frac{-E_\mathrm{mix}}{R} \left( \frac{1}{T_{\mathrm{ref}}} - \frac{1}{T} \right) \right) \\
K_{\mathrm{H_2}} &= k_{0,\mathrm{H_2}} \, \exp\left( \frac{-E_\mathrm{H_2}}{R} \left( \frac{1}{T_{\mathrm{ref}}} - \frac{1}{T} \right) \right)  \\
K_{\mathrm{OH}} &= k_{0,\mathrm{OH}} \, \exp\left( \frac{-E_\mathrm{OH}}{R} \left( \frac{1}{T_{\mathrm{ref}}} - \frac{1}{T} \right) \right) 
\end{align*}

The temperature dependence of the equilibrium constant $K_{\mathrm{eq}}$ can be described by the following approach:

\begin{equation*}
K_{\mathrm{eq}} = 137 \cdot \, 1.01325^{-2} \, T^{n_\mathrm{eq}} \, \, \exp\left( \frac{E_\mathrm{eq}}{R \, T}  \right) 
\end{equation*}

Assuming a gas velocity $v$ and a catalyst packing characterized by $\epsilon$ and $\rho_{\mathrm{Cat}}$, and using the reaction rate $r$, the concentration curves over the length coordinate $z$ of the reactor can then be described with the following ODE system:

\begin{align*}
\frac{\mathrm{d}c_\mathrm{CH_4}}{\mathrm{d}z} &= \nu_{\mathrm{CH4}} \, \frac{0.1}{v} \, \left( 1 - \epsilon \right) \, \rho_{\mathrm{Cat}} \, r \\
\frac{\mathrm{d}c_\mathrm{H_2O}}{\mathrm{d}z} &= \nu_{\mathrm{H_2O}} \, \frac{0.1}{v} \, \left( 1 - \epsilon \right) \, \rho_{\mathrm{Cat}} \, r \\
\frac{\mathrm{d}c_\mathrm{H_2}}{\mathrm{d}z}  &= \nu_{\mathrm{H_2}} \, \frac{0.1}{v} \, \left( 1 - \epsilon \right) \, \rho_{\mathrm{Cat}} \, r \\
\frac{\mathrm{d}c_\mathrm{CO_2}}{\mathrm{d}z} &= \nu_{\mathrm{CO_2}} \, \frac{0.1}{v} \, \left( 1 - \epsilon \right) \, \rho_{\mathrm{Cat}} \, r \\
\end{align*}

In order to obtain the partial pressures for the calculation of the reaction rate, the following definition is used:

\begin{equation*}
p_{\mathrm{i}} = \frac{c_{\mathrm{i}}}{\sum_{\mathrm{i}} c_{\mathrm{i}} } \, p = x_i \, p \quad \forall \, \mathrm{i} \in \{ \mathrm{CH_4}, \mathrm{H_2O}, \mathrm{H_2}, \mathrm{CO_2} \}
\end{equation*}

### Create a Dictionary for Parameters

```python
# Dictionary of parameters

para = {
    
    # Inlet Concentrations
    'cCO2_0' : 1,             # mol/m^3
    'cH2_0'  : 4,             # mol/m^3
    'cCH4_0' : 0,             # mol/m^3
    'cH2O_0' : 0,             # mol/m^3

    # Stoichiometric Coefficients
    'nu_CO2' : -1,            # -
    'nu_H2'  : -4,            # -
    'nu_CH4' :  1,            # -
    'nu_H2O' :  2,            # -
    
    # Kinetic Parameters
    'k0'     :  3.46e-4,      # mol/(g * s * bar)
    'k0_mix' :  0.88,         # bar^-0.5
    'k0_H2'  :  0.44,         # bar^-0.5
    'k0_OH'  :  0.50,         # bar^-0.5
    'E_a'    :  77500,        # J/mol
    'E_mix'  :  10000,        # J/mol
    'E_H2'   :  6200,         # J/mol
    'E_OH'   : -22400,        # J/mol
    'E_eq'   :  158700,       # J/mol
    'n_eq'   : -3.998,        # -

    # Model Parameters
    'R'      : 8.314,         # J/(mol * K)
    'epsil'  : 0.4,           # -
    'p'      : 10,            # bar
    'T_ref'  : 555,           # K
    'rho_cat': 2.3e6,         # g/m^3
    
    # Reactor Parameters
    'L'      : 1,             # m
    'T'      : 550,           # K
    'v'      : 1,             # m/s
}

# Concentration vector
para['c0'] = np.array([para['cCO2_0'], para['cH2_0'], para['cCH4_0'], para['cH2O_0']])
```

### Auxiliary functions ODE-System

```python
# Arrhenius equation function
def k(k0, E, para):
    '''
    Calculates the reaction rate constant using the Arrhenius equation
    
    Inputs:
    k0   : Pre-exponential factor
    E    : Activation energy
    para : Dictionary of parameters
    
    Output:
    k    : Reaction rate constant
    '''
    
    # Reaction rate constant
    k = k0 * np.exp((E / (para['R']) * (1/para['T_ref'] - 1/para['T'])))
    
    return k

# ODE System
def ODE_system(z,c,para):
    '''
    Solves the ODE system for the given concentrations, lenth of the reactor and the parameters
    
    Inputs:
    z    : Length of the reactor (Integration area) in m
    c    : Concentration vector in mol/m^3
    para : Dictionary of parameters
    
    Output:
    dcdz : Concentration gradients in mol/(m^3 * m)
    '''
    
    # Unpacking Conecetration Vector
    c_CO2   = c[0]              # mol/m^3
    c_H2    = c[1]              # mol/m^3
    c_CH4   = c[2]              # mol/m^3
    c_H2O   = c[3]              # mol/m^3
    
    # Unpacking Parameters
    epsil   = para['epsil']     # -
    rho_cat = para['rho_cat']   # g/m^3
    v       = para['v']         # m/s
    nu_CO2  = para['nu_CO2']    # -
    nu_H2   = para['nu_H2']     # -
    nu_CH4  = para['nu_CH4']    # -
    nu_H2O  = para['nu_H2O']    # -
    p       = para['p']         # bar  
    k0      = para['k0']        # mol/(g * s * bar)
    E_a     = para['E_a']       # J/mol
    k0_mix  = para['k0_mix']    # bar^-0.5
    E_mix   = para['E_mix']     # J/mol
    k0_H2   = para['k0_H2']     # bar^-0.5
    E_H2    = para['E_H2']      # J/mol
    k0_OH   = para['k0_OH']     # bar^-0.5
    E_OH    = para['E_OH']      # J/mol
    T       = para['T']         # K
    n_eq    = para['n_eq']      # -
    E_eq    = para['E_eq']      # J/mol
    R       = para['R']         # J/(mol * K)
    
    # Calculation of Kinetic Parameters
    K       = k(k0, E_a, para)                                                  # mol/(g * s)
    K_mix   = k(k0_mix, -E_mix, para)                                           # bar^-0.5
    K_H2    = k(k0_H2, -E_H2, para)                                             # bar^-0.5
    K_OH    = k(k0_OH, -E_OH, para)                                             # bar^-0.5
    K_eq    = 137 * (1.01325**-2) * np.power(T,n_eq) * np.exp(E_eq / (R * T))   # bar^-1
    
    # Calculation of Partial Pressures of the components
    p_CO2   = (c_CO2 / np.sum(c)) * p   # bar
    p_H2    = (c_H2  / np.sum(c)) * p   # bar
    p_CH4   = (c_CH4 / np.sum(c)) * p   # bar
    p_H2O   = (c_H2O / np.sum(c)) * p   # bar
    
    # Calculation of Reaction Rate
    Q       = 1 + (K_mix * (p_CO2**0.5)) + (K_H2 * (p_H2**0.5)) + (K_OH * p_H2O / (p_H2**0.5))                  # -
    r       = K * (p_CO2**0.5) * (p_H2**0.5) * (1 - ((p_CH4 * (p_H2O**2)) / (p_CO2 * (p_H2**4) * K_eq))) / Q**2 # mol/(g * s)
    
    # Calculation of the concentration gradients
    dCO2dz  = nu_CO2 * (0.1/v) * (1-epsil) * rho_cat * r    # mol/(m^3 * m)
    dH2dz   = nu_H2  * (0.1/v) * (1-epsil) * rho_cat * r    # mol/(m^3 * m)
    dCH4dz  = nu_CH4 * (0.1/v) * (1-epsil) * rho_cat * r    # mol/(m^3 * m)
    dH2Odz  = nu_H2O * (0.1/v) * (1-epsil) * rho_cat * r    # mol/(m^3 * m)
    
    # Merging the Concentration gradients into a single array
    dcdz    = np.array([dCO2dz, dH2dz, dCH4dz, dH2Odz])
    
    return dcdz
```

### Solving the ODE-System

```python
# Solve the system of ODEs

# Initial Conditions (Initial Concentrations)
c = para['c0']

# Integration area (Length of the reactor)
z = (0, para['L'])

# Define lambda-function
ODE_function = lambda z, c: ODE_system(z, c, para)

# Solving the ODE-System for Initial Conditions
sol = solve_ivp(ODE_function, z, c, method='DOP853', rtol=1e-6, t_eval=np.linspace(0,para['L'],100))
```

### Formulation of the Optimization Problem

The objective function is $F(T,v)$:
\begin{equation}
F(T,v) = -J(T,v)= -j_{\mathrm{CH_4}}(T,v) = -\Big(v \ c_\mathrm{CH_4}(T,v,z=L=1\ m)\Big)
\end{equation}

The formulated optimisation problem:

\begin{align*}
\max_{T,v\ \in\ \mathbb{R}} &\quad F(T,v) = -\min_{T,v\ \in\ \mathbb{R}} \Big( -J(T,v) \Big)= -\min_{T,v\ \in\ \mathbb{R}} \Big( -\big(v \ c_\mathrm{CH_4}(T,v,z=L=1\ m)\big) \Big) \\
\text{subject to} &\quad 0\ m \leq z \leq 1\ m \\
&\quad T \geq 550\ K \\
&\quad v \geq 1\ \frac{m}{s} \\
&\quad x_{CH_4}(z=L=1\ m) \leq 1 \%\\
&\quad \frac{\mathrm{d}c_i}{\mathrm{d}z} - \nu_i \, \frac{0.1}{v} \, \left( 1 - \epsilon \right) \, \rho_{\mathrm{Cat}} \, r(T,v) = 0 \quad \forall i: i \in \{CO_2, H_2, CH_4, H_2O \}\ \text{i.e. No accumulation} \\
                        &\quad c_i(z=0) = c_{i,in} \,  \, \, \, \qquad \qquad \, \, \, \qquad \, \, \, \qquad \forall i: i \in \{CO_2, H_2, CH_4, H_2O \}\ \text{i.e. initial concentrations are the inlet concentrations} \\
\end{align*}

### Auxiliary functions Optimization

```python
# Define new parameters dictionary for task b (Optimization of the methane output rate by varying the temperature and the gas velocity)
para_b = para.copy()

# Define the objective function to maximize the methane output
def objective_function_b(x, para_b, sign = -1.0):
    '''
    Objective function for the maximization of the methane production rate
    
    Input:
    x:      Array of optimization variables (T, v)
    para_b: Dictionary of parameters
    sign:   Sign of the objective function [Maximization(-1.0) or Minimization(1.0)]
    
    Output:
    jCH4:   Output of methane in mol/(m^2 s)
    '''
    
    # Include search point in the dictionoary
    # Temperature
    para_b['T'] = x[0]
    # Gas velocity
    para_b['v'] = x[1]
    
    # Initial Conditions
    c    = para_b['c0']

    # Integration area
    z    = (0, para_b['L'])

    # Define lambda-function
    ODE_function_b = lambda z, c: ODE_system(z, c, para_b)

    # Solving the ODE-System for Initial Conditions
    sol_b  = solve_ivp(ODE_function_b, z, c, method='DOP853', rtol=1e-6, t_eval=np.linspace(0,para_b['L'],100))

    # Unpacking the Trajectory of the Concentration of methane
    cCH4 = sol_b.y[2, :]
    
    # Calculate the methane output
    jCH4 = para_b['v'] * cCH4[-1]
    
    # Return the value of the methane output at the end of the reactor
    return sign * jCH4
```

```python
# Reset the parameters to the initial values
para_b['T'] = 550   # K
para_b['v'] = 1     # m/s

# Define the bounds
T_bound = (550, None)  # Temperature Constraint (T >= 550 K)
v_bound = (1, None)    # Gas velocity Constraint (v >= 1 m/s)
bounds_b = (T_bound, v_bound)

# Define the constraints
def constraint_b(x,para_b):
    '''
    Constraint function for the CO2 composition to be less than 1% at the end of the reactor
    
    Input:
    x:      Array of optimization variables (T, v)
    para_b: Dictionary of parameters
    
    Output:
    1 - xCO2[-1]:   CO2 composition limit to maximum of 1%
    '''
    # Temperature
    para_b['T'] = x[0]
    # Gas velocity
    para_b['v'] = x[1]
    
    # Initial Conditions
    c = para_b['c0']

    # Integration area
    z = (0, para_b['L'])

    # Define lambda-function
    ODE_function_b = lambda z, c: ODE_system(z, c, para_b)

    # Solving the ODE-System for Initial Conditions
    sol_b = solve_ivp(ODE_function_b, z, c, method='DOP853', rtol=1e-6, t_eval=np.linspace(0,para_b['L'],100))
    
    # Unpacking the Trajectories of the Concentrations
    cCO2 = sol_b.y[0, :]
    cH2  = sol_b.y[1, :]
    cCH4 = sol_b.y[2, :]
    cH2O = sol_b.y[3, :]
    
    # Calculation of Mole Fraction of CO2
    xCO2 = (cCO2 / (cCO2 + cH2 + cCH4 + cH2O)) * 100
    
    # Return the constraint for CO2 conversion (X_CO2 <= 1%)
    return 1 - xCO2[-1]

# Define the inequality constraints as a list of dictionaries
cons_b = [{'type': 'ineq', 'fun': lambda x: constraint_b(x,para_b)}]

# Define the initial guess
x0_b = np.array([para_b['T'], para_b['v']])

# Solve the optimization problem
obj_fun_min_b = lambda x : objective_function_b(x,para_b)
res_b = minimize(obj_fun_min_b, x0_b, method='SLSQP', constraints=cons_b, bounds=bounds_b, options={'disp': True})

# Unpacking the results
T_opt_b = res_b.x[0]    # Optimal Temperature
v_opt_b = res_b.x[1]    # Optimal Gas Velocity
jCH4_opt_b = -res_b.fun # Optimal Methane Output

# Reset the parameters to the optimal values
# Temperature
para_b['T'] = T_opt_b
# Gas velocity
para_b['v'] = v_opt_b

# Calculate Concentration and Mole Fractions for optimized parameters
# Initial Conditions
c    = para_b['c0']

# Integration area
z    = (0, para_b['L'])

# Define lambda-function
ODE_function_b = lambda z, c: ODE_system(z, c, para_b)

# Solving the ODE-System for Initial Conditions
sol_b  = solve_ivp(ODE_function_b, z, c, method='DOP853', rtol=1e-6, t_eval=np.linspace(0,para_b['L'],100))

# Unpacking the Trajectories of the Concentrations
b_cCO2 = sol_b.y[0, :]
b_cH2  = sol_b.y[1, :]
b_cCH4 = sol_b.y[2, :]
b_cH2O = sol_b.y[3, :]

# Calculation of Mole Fractions
b_xCO2 = (b_cCO2 / (b_cCO2 + b_cH2 + b_cCH4 + b_cH2O)) * 100
b_xH2  = (b_cH2  / (b_cCO2 + b_cH2 + b_cCH4 + b_cH2O)) * 100
b_xCH4 = (b_cCH4 / (b_cCO2 + b_cH2 + b_cCH4 + b_cH2O)) * 100
b_xH2O = (b_cH2O / (b_cCO2 + b_cH2 + b_cCH4 + b_cH2O)) * 100
```

### Selection of a third optimization parameter

There are two parameters that can be optimized in order to obtain even higher methane output:
- **Reactor Length (L)**: The longer the reactor, the higher the reaction time. So, by increasing the reactor length, methane output can be increased.
- **Reactor Pressure (p)**: Pressure is directly related to the adsorption of the gases on to the catalyst and hence the reaction rate. So, by increasing the pressure, reaction rate can be increased and ultimately methane output can be increased.

But, here we can only optimitze the reaction parameters, and changing the reactor pressure is much easier than the reactor length (physically), so here **reactor pressure (p)** is chosen as the optimization parameter rather than the **reactor length (L)**.
